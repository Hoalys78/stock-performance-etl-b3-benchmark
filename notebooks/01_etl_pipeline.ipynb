# Stock Performance ETL - Portfolio Notebook

This notebook demonstrates the ETL pipeline for B3 stocks + benchmarks.
Clean, documented version for GitHub portfolio.


```python
from src.config import TICKERS, START_DATE, END_DATE
from src.extract import extract_quotes
from src.transform import transform_data
from src.dimensions import create_dim_asset, create_dim_calendar
from src.load import export_data
from src.quality import data_quality_report

import pandas as pd
import matplotlib.pyplot as plt
```

```python
# 1. EXTRACT
raw_df = extract_quotes(TICKERS, START_DATE, END_DATE)
raw_df.head()
```

```python
# 2. TRANSFORM
fact_prices = transform_data(raw_df)
fact_prices.head()
```

```python
# 3. DIMENSIONS
dim_asset = create_dim_asset(fact_prices)
dim_calendar = create_dim_calendar(fact_prices)
dim_asset
```

```python
# 4. QUALITY REPORT
data_quality_report(fact_prices)
```

```python
# 5. QUICK VIZ - Cumulative Return
import pandas as pd
pivot = fact_prices.pivot(index='date', columns='ticker', values='close_price')
returns = (pivot / pivot.iloc[0] - 1) * 100
returns.plot(figsize=(12,6), title='Cumulative Return % Since 2022')
plt.ylabel('Return %')
plt.grid(alpha=0.3)
plt.show()
```

```python
# 6. EXPORT
export_data(fact_prices, dim_asset, dim_calendar)
```
